```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import tkinter as tk
import joblib

# Load the model
forest_clf = joblib.load('random_forest_model.joblib')

def predict():
    # Create a pandas DataFrame for our input features
    input_data = pd.DataFrame([[
        float(tempmax_entry.get()), 
        float(tempmin_entry.get()), 
        float(temp_entry.get()),
        float(feelslike_entry.get()), 
        float(precip_entry.get()),
        float(humidity_entry.get()), 
        float(windspeed_entry.get())
    ]], columns=['tempmax', 'tempmin', 'temp', 'feelslike', 'precip', 'humidity', 'windspeed'])
    
    # Use the model to predict the target
    prediction = forest_clf.predict(input_data)
    result_label.config(text = "Prediction result: " + str(prediction[0]))

# Create a tkinter window
window = tk.Tk()
window.title("Weather Prediction")

tk.Label(window, text="Max Temperature").pack()
tempmax_entry = tk.Entry(window)
tempmax_entry.pack()

tk.Label(window, text="Min Temperature").pack()
tempmin_entry = tk.Entry(window)
tempmin_entry.pack()

tk.Label(window, text="Temperature").pack()
temp_entry = tk.Entry(window)
temp_entry.pack()

tk.Label(window, text="Feels Like").pack()
feelslike_entry = tk.Entry(window)
feelslike_entry.pack()

tk.Label(window, text="Precipitation").pack()
precip_entry = tk.Entry(window)
precip_entry.pack()

tk.Label(window, text="Humidity").pack()
humidity_entry = tk.Entry(window)
humidity_entry.pack()

tk.Label(window, text="Wind Speed").pack()
windspeed_entry = tk.Entry(window)
windspeed_entry.pack()

predict_button = tk.Button(window, text="Predict", command=predict)
predict_button.pack()

result_label = tk.Label(window, text="Prediction result will appear here", font=("Arial", 18))
result_label.pack()

window.mainloop()
```
